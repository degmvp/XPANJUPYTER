```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
def graf():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x, y, z = axes3d.get_test_data(0.05)
    ax.plot_wireframe(x,y,z, rstride=2, cstride=2)

    plt.show()
```

```python
graf()
```
