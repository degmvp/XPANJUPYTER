```python
from csv import reader
from matplotlib import pyplot as plt

with open('M://area_matplotlib//objetos.txt') as file:
    parse = reader(file)

    data_2019 = list(filter(lambda v: v[0] == '2019', parse))

    max_temp = [float(v[3]) for v in data_2019 if v[3]]
    min_temp = [float(v[4]) for v in data_2019 if v[4]]
    med_temp = [float(v[5]) for v in data_2019 if v[5]]

    # import pdb; pdb.set_trace()

    plt.plot(max_temp, label='TOT-PROCS')
    plt.plot(min_temp, label='TOT-TABLES')
    plt.plot(med_temp, label='TOT-VIEWS')

    plt.legend()
    plt.xlabel('Dias do ano')
    plt.ylabel('Objetos DB')
    plt.title('Movimentação de Objetos XPYDB')
    plt.show()
```
